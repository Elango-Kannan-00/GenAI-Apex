```python
from transformers import pipeline
```

```python
import pandas as pd
```

```python
data=pd.read_excel("/content/Employee Sample Data.xlsx",header=0)
```

```python
data=data.dropna()
```

```python
data
```

```python
from transformers import AutoTokenizer, AutoModelForTableQuestionAnswering
```

```python
model='google/tapas-large-finetuned-wtq'
```

```python
tokenizer=AutoTokenizer.from_pretrained(model)
```

```python
tokenizer
```

```python
model=AutoModelForTableQuestionAnswering.from_pretrained(model)
```

```python
tqapipe=pipeline(task="table-question-answering",model=model,tokenizer=tokenizer)
```

```python
tqapipe(table=data,query="tell me the name of E00549")
```
